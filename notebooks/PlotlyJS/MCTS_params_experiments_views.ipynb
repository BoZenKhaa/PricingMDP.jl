```julia
using DataFrames
using CSV
using PlotlyJS
```

```julia
dataset(DataFrame, "iris");
```

```julia
res = CSV.File("mcts_params_results.csv") |> DataFrame;
# res = res[1:5, :]
```

```julia
df = res
x = :exploration_constant
y= :n_iterations 
z= :depth
marker_color = :mean_r
```

```julia
plot(df, Layout(title = "MCTS parameters",
                margin=attr(l=0, r=0, b=0, t=0), 
                scene = attr(
                    xaxis_title="$(x)",
                    yaxis_title="$(y)",
                    zaxis_title="$(z)"),
                coloraxis=attr(colorscale="Viridis",),
        ),
    x=x, y=y, z=z, marker_color = marker_color, text = marker_color, # text -> adds hover info about value
    type="scatter3d", mode="markers",
    marker = attr(
                size=10,
                colorscale="Viridis",   # choose a colorscale
                opacity=0.8, 
                showscale=true
            )
)
```
