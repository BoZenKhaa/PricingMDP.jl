```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path
from Python.pricingmdprunner.utils import load_yaml
```

```python
experiments_path = Path(r"C:\Users\mrkos\scth\projects\MDPPricing\data\ev_mcts_analysis\single_day_cs_pp_T=96_c=3_expected_res=12_nᵣ=12_res_budget_μ=2.0\results\mcts")

df = pd.DataFrame(columns=['depth', 'exploration_constant', 'n_iterations', 'reuse_tree', 'r_mean'])

for config_path in experiments_path.rglob("**/config*.yaml"):
    config = load_yaml(config_path)
    result_path = config_path.parent / f"{config_path.stem}_result.csv"
    result = pd.read_csv(result_path)
    
    cfg = config["solver_params"]
    cfg.pop("rng")
    entry = pd.DataFrame.from_dict([cfg,])
    entry['r_mean'] = result.r.mean()
    
    df = pd.concat([df, entry], ignore_index=True)
    # break
```
